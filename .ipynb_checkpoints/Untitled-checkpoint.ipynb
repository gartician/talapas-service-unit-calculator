```python
import os
import sys
import re
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import dash_bootstrap_components as dbc
import plotly.graph_objs as go
import plotly.express as px
import pandas as pd
import itertools
```

```python
su = 140
max_times = 10
su_dollars = 0.025

def cost_surface(est_cost, units = "su"):
    number_d = pd.Series(range(1, max_times))
    frequency = pd.Series(range(1, max_times))
    df_surface = pd.DataFrame(columns=['Number of Days', 'Frequency'])
    for i in itertools.product(pd.Series(range(1, 10)), pd.Series(range(1, 10))):
        temp_row = pd.Series(list(i), index=['Number of Days', 'Frequency'])
        df_surface = df_surface.append(temp_row, ignore_index=True)
    if units == "su":
        df_surface = df_surface.assign(Cost = df_surface['Number of Days'] * df_surface['Frequency'] * su)
    elif units == "dollars":
        df_surface = df_surface.assign(Cost = df_surface['Number of Days'] * df_surface['Frequency'] * su * su_dollars)
    else:
        print("incorrect unit type. Must be 'su' or 'dollars'. ")
    return(df_surface)

cost_surface(su, units="dollars")
    
# def cost_table(est_cost, units = "su"):
#     df = pd.DataFrame(columns=["Number_of_Days", "Frequency"])
#     df['Number_of_Days'] = pd.Series(range(1, max_times))
#     df['Frequency'] = pd.Series(range(1, max_times))
#     if units == "su":
#         df = df.assign(cost = df['Number_of_Days'] * df['Frequency'] * est_cost)
#         df.rename(columns={'cost': "Cost_SU"}, inplace=True)
#         return(df)
#     elif units == "dollars":
#         df = df.assign(cost = df['Number_of_Days'] * df['Frequency'] * est_cost * 0.025)
#         df.rename(columns={'cost': "Cost_Dollars"}, inplace=True)
#         return(df)

# cost_table(1000)
# cost_table(su, "dollars")
```
